## Trabalhando com planilhas do Excel - Bootcamp Unimed-BH

```python
# Carregando arquivos
df1 = pd.read_excel('./dados/Aracaju.xlsx')
df2 = pd.read_excel('./dados/Fortaleza.xlsx')
df3 = pd.read_excel('./dados/Natal.xlsx')
df4 = pd.read_excel('./dados/Recife.xlsx')
df5 = pd.read_excel('./dados/Salvador.xlsx')
```

```python
# Concatenando os dados
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
# Visualizando os dados iniciais
df.head(10)
```

```python
# Visualizando os dados finais
df.tail(10)
```

```python
# Visualizando os tipos de dados
df.dtypes
```

```python
# Alterando tipo de dados da coluna
df['LojaID'] = df['LojaID'].astype('object')
```

```python
# Visualizando os tipos de dados
df.dtypes
```

```python
# Verificando os valores nulos
df.isnull().sum()
```

```python
# Verificando se tem valores nulos
df.sample(20)
```

```python
## Formas para tratar valores nulos

# Substituindo os valores nulos por zero
df['Vendas'].fillna(0, inplace=True)

# Apagando as linhas com valores nulos
df.dropna(inplace=True)

# Apagando as linhas com valores nulos com base apenas em uma coluna
df.dropna(subset=['Vendas'], inplace=True)

# Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how='all', inplace=True)
```

```python
# Criando coluna de receita
df['Receita'] = df['Vendas'].mul(df['Qtde'])
```

```python
# Visualizando os dados
df.head()
```

```python
# Visualizando a receita por loja
df.groupby('LojaID')['Receita'].sum().sort_values()
```

```python
# Visualizando a receita por cidade
df.groupby('Cidade')['Receita'].sum().sort_values()
```

## Trabalhando com datas - Bootcamp Unimed-BH

```python
# Supondo que os dados vieram com a coluna de data sendo int ou float

# Convertendo a coluna data
df['Data'] = pd.to_datetime(df['Data'])  # Assim convertemos os valores da coluna data para data
```

```python
# Precisamos saber qual é a receita por ano
df.groupby(df['Data'].dt.year)['Receita'].sum()
```

```python
# Criando uma nova coluna com o ano
df['Ano_Venda'] = df['Data'].dt.year
```

```python
# Visualizando os dados
df.sample(5)
```

```python
# Extraindo o mês e o dia da venda
df['Mes_Venda'], df['Dia_Venda'] = (df['Data'].dt.month, df['Data'].dt.day)
```

```python
# Visualizando os dados
df.sample(5)
```

```python
# Retornando a data mais antiga
df['Data'].min()
```

```python
# Calculando a diferença de dias
df['Diferenca_Dias'] = df['Data'] - df['Data'].min()
```

```python
# Visualizando dados
df.sample(6)
```

```python
# Criando a coluna de trimestre
df['Trimestre_Venda'] = df['Data'].dt.quarter
```

```python
# Visualizando os dados
df.sample(10)
```

```python
# Filtrando as vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df['Data'].dt.year == 2019) & (df['Data'].dt.month == 3)]
```

```python
# Visualizando os dados
vendas_marco_19.sample(20)
```

## Visualização de dados - Bootcamp Unimed-BH

```python
# Retornando a quantidade de vendas por lojas
df['LojaID'].value_counts()
```

```python
# Visualizando os dados em grafico
df['LojaID'].value_counts().plot.bar();
```

```python
# Visualizando os dados em grafico
df['LojaID'].value_counts(ascending=True).plot.barh();
```

```python
# Precisamos saber qual é a receita por ano, visualizando os dados em grafico
df.groupby(df['Data'].dt.year)['Receita'].sum().plot.pie();
```

```python
# Total de vendas por cidade
df['Cidade'].value_counts()
```

```python
# Total de vendas por cidade
df['Cidade'].value_counts().plot.bar(title='Total vendas por cidade')
plt.xlabel('Cidade')
plt.ylabel('Total Vendas')
```

```python
# Mudando o estilo do grafico
plt.style.use('ggplot')
```

```python
# Visualizando grafico de linha
df.groupby(df['Mes_Venda'])['Qtde'].sum().plot(title='Total Produtos Vendidos por Mês')
plt.xlabel('Mês')
plt.ylabel('Total Produtos Vendidos');
plt.legend()
```

```python
# Retornando a quantidade de vendas por mês
df.groupby(df['Mes_Venda'])['Qtde'].sum()
```

```python
# Selecionando apenas as vendas de 2019
df_2019 = df[df['Ano_Venda'] == 2019]
```

```python
# Total produtos vendidos por mês
df_2019.groupby(df_2019['Mes_Venda'])['Qtde'].sum().plot(marker = 'o')
plt.xlabel('Mês')
plt.ylabel('Total Produtos Vendidos');
plt.legend()
```

```python
# Histograma
plt.hist(df['Qtde']);
```

```python
# Visualizando um grafico de dispersão
plt.scatter(x=df_2019['Dia_Venda'], y=df_2019['Receita']);
```

```python
# Salvando o grafico em png
df_2019.groupby(df_2019['Mes_Venda'])['Qtde'].sum().plot(marker = 'v')
plt.title('Quantidade de Produtos Vendidos x Mês')
plt.xlabel('Mês')
plt.ylabel('Total Produtos Vendidos');
plt.legend()
plt.savefig('Grafico Qtde x Mês.png')
```
